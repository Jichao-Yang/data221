```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/final/wrangled/pca_clusters.csv')
df = df[['instability', 'kmeans_labels', 'year']]
```

```python
import seaborn as sns

# Get unique kmeans_labels
unique_labels = df['kmeans_labels'].unique()

# Plot KDEs
plt.figure(figsize=(10, 6))
for label in unique_labels:
    subset = df[df['kmeans_labels'] == label]
    sns.kdeplot(subset['year'], label=f'Label {label}', shade=True)

plt.xlabel('Year')
plt.ylabel('Density')
plt.legend()
plt.show()
```
